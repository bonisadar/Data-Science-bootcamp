## Cosine Similarity Calculations
Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them. Similarity measures have a multiude of uses in machine learning projects; they come in handy when matching strings, measuring distance, and extracting features. This similarity measurement is particularly concerned with orientation, rather than magnitude. 
In this case study, you'll use the cosine similarity to compare both a numeric data within a plane and a text dataset for string matching.

Load the Python modules, including cosine_similarity, from sklearn.metrics.pairwise

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
plt.style.use('ggplot')
from scipy import spatial
from sklearn.metrics.pairwise import cosine_similarity
```

**<font color='teal'> Load the distance dataset into a dataframe. </font>**

```python
df = pd.read_csv('distance_dataset.csv', index_col=0)
df.head()
```

### Cosine Similarity with clusters and numeric matrices

All points in our dataset can be thought of as feature vectors. We illustrate it here as we display the __Cosine Similarity__ between each feature vector in the YZ plane and the [5, 5] vector we chose as reference. The sklearn.metrics.pairwise module provides an efficient way to compute the __cosine_similarity__ for large arrays from which we can compute the similarity.

 **<font color='teal'> First, create a 2D and a 3D matrix from the dataframe. The 2D matrix should contain the 'Y' and 'Z' columns and the 3D matrix should contain the 'X','Y', and 'Z' columns.</font>**

```python
matYZ = df[['Y', 'Z']].to_numpy()
mat = df[['X', 'Y', 'Z']].to_numpy()
```

Calculate the cosine similarity for those matrices with reference planes of 5,5 and 5,5,5. Then subtract those measures from 1 in new features.

```python
simCosine3D = 1. - cosine_similarity(mat, [[5,5,5]])
simCosine = 1. - cosine_similarity(matYZ, [[5,5]])
```

Using the 2D matrix and the reference plane of (5,5) we can use a scatter plot to view the way the similarity is calculated using the Cosine angle.

```python
figCosine = plt.figure(figsize=[10,8])

plt.scatter(df.Y, df.Z, c=simCosine[:,0], s=20)
plt.plot([0,5],[0,5], '--', color='dimgray')
plt.plot([0,3],[0,7.2], '--', color='dimgray')
plt.text(0.7,2.6,r'$\theta$ = 22.4 deg.', rotation=68, size=14)
plt.ylim([0,10])
plt.xlim([0,10])
plt.xlabel('Y', size=14)
plt.ylabel('Z', size=14)
plt.title('Cosine Similarity')
cb = plt.colorbar()
cb.set_label('Similarity with (5,5)', size=14)

#figCosine.savefig('similarity-cosine.png')
```

Now, plot the 3D matrix with the similarity and the reference plane, (5,5,5).

```python
from mpl_toolkits.mplot3d import Axes3D
figCosine3D = plt.figure(figsize=(10, 8))
ax = figCosine3D.add_subplot(111, projection='3d')

p = ax.scatter(mat[:,0], mat[:,1], mat[:,2], c=simCosine3D[:,0])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
cb = figCosine3D.colorbar(p)
cb.set_label('Similarity with (5,5,5)', size=14)
                                   
figCosine3D.tight_layout()
#figCosine3D.savefig('cosine-3D.png', dpi=300, transparent=True)
```

### Cosine Similarity with text data
This is a quick example of how you can use Cosine Similarity to compare different text values or names for record matching or other natural language proecessing needs. 
First, we use count vectorizer to create a vector for each unique word in our Document 0 and Document 1. 

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
Document1 = "Starbucks Coffee"
Document2 = "Essence of Coffee"

corpus = [Document1,Document2]

X_train_counts = count_vect.fit_transform(corpus)

pd.DataFrame(X_train_counts.toarray(),columns=count_vect.get_feature_names_out(),index=['Document 0','Document 1'])
```

Now, we use a common frequency tool called TF-IDF to convert the vectors to unique measures.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

trsfm=vectorizer.fit_transform(corpus)

pd.DataFrame(trsfm.toarray(),columns=vectorizer.get_feature_names_out(),index=['Document 0','Document 1'])

# Calculating Cosine Similarity between Document 0 and Document 1
cosine_sim = cosine_similarity(trsfm[0:1], trsfm[1:2])

print("\nCosine Similarity between Document 0 and Document 1:", cosine_sim[0][0])
```

Here, we finally apply the __Cosine Similarity__ measure to calculate how similar Document 0 is compared to any other document in the corpus. Therefore, the first value of 1 is showing that the Document 0 is 100% similar to Document 0 and 0.26055576 is the similarity measure between Document 0 and Document 1.

```python
cosine_similarity(trsfm[0:1], trsfm)
```

### Lemmatization and Stopword Removal

```python
import spacy
import string

# Loading spaCy's English language model
#The model en_core_web_sm is a small pre-trained model for English, containing information about language syntax, parts of speech (POS), 
#named entities, and more. This model is used by spaCy for processing natural language.
#The "sm" version is commonly used for simpler tasks.
nlp = spacy.load('en_core_web_sm')

def preprocess(text):
    '''A Function for preprocessing: Lemmatization and Stopword Removal'''
    # Tokenize, remove stopwords and punctuation, and lemmatize the text
    doc = nlp(text)  # This converts the input text into a spaCy document object. When a text string is passed to nlp(), 
                     # spaCy tokenizes the string into individual words (called "tokens") and performs various language processing tasks 
                     # such as part-of-speech tagging, dependency parsing, and named entity recognition (NER).
                     # The resulting doc object contains all the tokens in the input text along with their linguistic features (like lemma, POS, etc.).
    
    return ' '.join([token.lemma_ for token in doc if not token.is_stop and token.text not in string.punctuation])
    # token.lemma_ returns the lemma of the token. The lemma is the base form of a word (i.e., the dictionary form). For example:
    # "running" → "run" "better" → "good" "learned" → "learn" 

    # Lemmatization is the process of reducing words to their base or root form, which helps in consolidating different forms of a word 
    # (e.g., "runs", "ran", "running") to a single form ("run").

Document1 = "Learning Data Science from online resources"
Document2 = "Using online resources to learn Data Science"

# Preprocess the documents
corpus = [preprocess(Document1), preprocess(Document2)]

vectorizer = TfidfVectorizer()

trsfm = vectorizer.fit_transform(corpus)

df = pd.DataFrame(trsfm.toarray(), columns=vectorizer.get_feature_names_out(), index=['Document 0', 'Document 1'])
print("TF-IDF Matrix:\n", df)

# Calculating Cosine Similarity between Document 0 and Document 1
cosine_sim = cosine_similarity(trsfm[0:1], trsfm[1:2])

print("\nCosine Similarity between Document 0 and Document 1:", cosine_sim[0][0])
```

## Using Word Embeddings

```python
from gensim.models import Word2Vec

Document1 = "Learning Data Science from online resources"
Document2 = "Using online resources to learn Data Science"

# Tokenize the documents
doc1_tokens = Document1.lower().split()
doc2_tokens = Document2.lower().split()

# Trainning a simple Word2Vec model
model = Word2Vec([doc1_tokens, doc2_tokens], vector_size=50, window=5, min_count=1, workers=4)

def avg_word2vec(tokens, model):
    '''A Function to average word vectors for a document'''
    vectors = [model.wv[token] for token in tokens if token in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

# Getting the average Word2Vec vectors for the two documents
vec1 = avg_word2vec(doc1_tokens, model)
vec2 = avg_word2vec(doc2_tokens, model)

cosine_sim = cosine_similarity([vec1], [vec2])
print("Cosine Similarity between Document 1 and Document 2 using Word2Vec:", cosine_sim[0][0])
```

## Using Pre-trained BERT Models

The BERT (Bidirectional Encoder Representations from Transformers) model was developed by Google AI and trained on massive amounts of text data, including the English Wikipedia and BooksCorpus (a dataset containing 11,000 books). It was trained on these corpora using a self-supervised learning task called Masked Language Modeling (MLM) and Next Sentence Prediction (NSP). This means that the model was trained to predict missing words in sentences and to understand relationships between sentence pairs.

```python
from transformers import BertTokenizer, BertModel
import torch

# Loading pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

Document1 = "Learning Data Science from online resources"
Document2 = "Using online resources to learn Data Science"

# Tokenizing the documents
inputs1 = tokenizer(Document1, return_tensors='pt', truncation=True, padding=True, max_length=512)
inputs2 = tokenizer(Document2, return_tensors='pt', truncation=True, padding=True, max_length=512)

# tokenizer: The tokenizer converts the input text (documents in this case) into tokens that the BERT model can understand. 
# It breaks the sentences into smaller units like words or subwords.

# return_tensors='pt': This tells the tokenizer to return PyTorch tensors (necessary for input to the BERT model). 
# These tensors are the numerical representations of the tokens.

# truncation=True: This ensures that if a document is too long (more than max_length), it will be truncated to fit within the maximum token length.

# padding=True: This ensures that if a document has fewer tokens than the max_length, it will be padded with zeros to make it fit.

# max_length=512: The maximum number of tokens BERT can process at once is 512. This ensures that each input is either padded or truncated 
# to a length of 512 tokens.

# After this step, inputs1 and inputs2 are dictionaries that contain the tokenized representations of Document1 and Document2, 
# ready for the BERT model.


# Getting embeddings from BERT
with torch.no_grad():
    outputs1 = model(**inputs1)
    outputs2 = model(**inputs2)

# torch.no_grad(): This is a context manager that tells PyTorch not to track gradients (since we're not training the model but only 
# using it for inference). It saves memory and computation by preventing unnecessary gradient calculations.

# model(**inputs1) and model(**inputs2): These lines pass the tokenized documents into the pre-trained BERT model. The model processes the 
# inputs and returns several outputs, one of which is the hidden states (the embeddings of the tokens).


# Extracting embeddings (using the [CLS] token's embedding)
embeddings1 = outputs1.last_hidden_state[:, 0, :].numpy()
embeddings2 = outputs2.last_hidden_state[:, 0, :].numpy()

# outputs1.last_hidden_state: The BERT model returns several values, and last_hidden_state is the final output of the model (the hidden 
# states at each layer). It has a shape of (batch_size, sequence_length, hidden_size), 
# where:
# batch_size is the number of documents passed (in this case, 1 for each).
# sequence_length is the number of tokens in the document.
# hidden_size (dimensionality of the token embeddings) is the size of the vector representing each token (for bert-base-uncased, this is 768).

# [:, 0, :]: This selects the first token's embedding, which is the [CLS] token. The [CLS] token is used as a representation of the entire 
# document (or sentence) in BERT-based models.

# : (before the comma) means selecting all instances in the batch (in this case, there's just one document).

# 0 refers to the first token (the [CLS] token) in the sequence.

# : after the comma means selecting all features of that token (i.e., its entire embedding vector of size 768).

# .numpy(): This converts the PyTorch tensor into a NumPy array, which is easier to handle for similarity calculation.

# After this step, embeddings1 and embeddings2 contain the 768-dimensional vectors (embeddings) for each document’s [CLS] token, 
# which encapsulate the document's semantic representation.

# Calculating Cosine Similarity
cosine_sim = cosine_similarity(embeddings1, embeddings2)
print("Cosine Similarity between Document 1 and Document 2 using BERT:", cosine_sim[0][0])
```
