**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(*borough_names):
    '''Function to return the ratio of Avg. price of London boroughs for the year 1998 and 2018 as a DataFrame'''
    
    boroughs = []
    ratios = []
    
    for borough_name in borough_names:
        avg_1998 = properties[(properties['London_borough'] == borough_name) & (properties['Year'] == 1998)]['Average_price'].mean()
        avg_2018 = properties[(properties['London_borough'] == borough_name) & (properties['Year'] == 2018)]['Average_price'].mean()
        ratio_avg = avg_1998 / avg_2018
        
        boroughs.append(borough_name)
        ratios.append(ratio_avg)
        
    return pd.DataFrame({'London_borough': boroughs, 'Price_ratio':ratios})
```

```python
# Creating a df consisting the average price and borough name
all_ratios_df = create_price_ratio(*properties['London_borough'].unique())
```

```python
all_ratios_df.head()
```

```python
def create_price_2018(*borough_names):
    '''Function to return the average price of boroughs in 2018 as a DataFrame'''
    
    boroughs = []
    prices = []
    for borough_name in borough_names:
        avg_2018 = properties[(properties['London_borough'] == borough_name) & (properties['Year'] == 2018)]['Average_price'].mean()
        
        boroughs.append(borough_name)
        prices.append(avg_2018)
        
    return pd.DataFrame({'London_borough': boroughs, 'Prices_2018': prices})
```

```python
# Creating a df consisting the average price and borough name
price_2018_df = create_price_2018(*properties['London_borough'].unique())
```

```python
# Visualization of the property price ratio of London borough over two decade
sns.set_theme(style="white", context="talk")
plt.figure(figsize=(12, 8))

bar_plot = sns.barplot(x='London_borough', y='Price_ratio', data=all_ratios_df.sort_values('Price_ratio'), palette='mako', hue='London_borough', legend=False)

plt.title('Price Ratio for property values in 1998 vs 2018 for Different Boroughs', fontsize=16)
plt.xlabel('London Borough', fontsize=12)
plt.ylabel('Price Ratio (1998 / 2018)', fontsize=12)

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame by Price_ratio and get the top 3 lowest
sorted_boroughs = all_ratios_df.sort_values('Price_ratio').head(3)
sorted_boroughs
```

```python
#  Visualizing the property prices in 2018 for the top three London boroughs with the highest price increases.

sns.set_theme(style="white", context="talk")
plt.figure(figsize=(12, 8))

# Define a custom formatter for y-axis
def millions_formatter(x, pos):
    return f'{x * 1e-6:.1f}M'

# Creating a list of boroughs that reflects most property price increase
highlight = ['Hackney', 'Waltham Forest', 'Southwark']

# Define custom color palette
colors = ['red' if borough in highlight else 'lightskyblue' for borough in price_2018_df.sort_values('Prices_2018')['London_borough']]

bar_plot = sns.barplot(x='London_borough', y='Prices_2018', data=price_2018_df.sort_values('Prices_2018'), palette=colors, hue='London_borough', legend=False)

plt.title('Property Prices of London borough at 2018', fontsize=16)
plt.xlabel('London Borough', fontsize=12)
plt.ylabel('Prices_2018', fontsize=12)

plt.xticks(rotation=90)

# Format the y-axis
ax = plt.gca()  # Get the current axis
ax.yaxis.set_major_formatter(mtick.FuncFormatter(millions_formatter))

plt.tight_layout()
plt.show()
```
