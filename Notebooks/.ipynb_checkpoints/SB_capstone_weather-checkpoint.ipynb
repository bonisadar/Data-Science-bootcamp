```python
df_no_missing = df_csv.dropna()
```

```python
df_few = df_no_missing[df_no_missing['city_name'].isin(['Bismarck	', 'Helena', 'Saint Paul', 'Phoenix'])]
```

```python
df_few_sel_col = df_few.iloc[:, [0] + list(range(3, 10))]
df_few_sel_col
```

```python
df_new_index = df_few_sel_col.set_index('city_name')
```

```python
df_index = df_new_index.index
df_columns = df_new_index.columns
```

```python
df_scale = scale(df_new_index)
```

```python
df_df = pd.DataFrame(df_scale, columns=df_columns)
df_df
```

```python
df_df.mean()
```

```python
df_df.std(ddof=0)
```

```python
pca_df = PCA().fit(df_df)
```

```python
plt.subplots(figsize=(10, 6))
plt.plot(pca_df.explained_variance_ratio_.cumsum())
plt.xlabel('Component #')
plt.ylabel('Cumulative ratio variance')
plt.title('Cumulative variance ratio explained by PCA components for weather summary statistics');
```

```python
df_x = pca_df.transform(df_scale)
```
