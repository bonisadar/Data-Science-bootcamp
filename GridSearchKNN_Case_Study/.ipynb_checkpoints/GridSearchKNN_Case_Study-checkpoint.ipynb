## Grid Search Hyperparameter optimization

This case study is all about using grid searches to identify the optimal parameters for a machine learning algorithm. To complere this case study, you'll use the Pima Indian diabetes dataset from Kaggle and KNN. Follow along with the preprocessing steps of this case study.

Load the necessary packages

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# set random seed to try make this exercise and solutions reproducible (NB: this is just for teaching purpose and not something you would do in real life)
random_seed_number = 42
np.random.seed(random_seed_number)
```

#### Load the diabetes data

```python
diabetes_data = pd.read_csv('data/diabetes.csv')
diabetes_data.head()
```

**<font color='teal'> Start by reviewing the data info.</font>**

```python
diabetes_data.info()
```

**<font color='teal'> Apply the describe function to the data.</font>**

```python
diabetes_data.describe()
```

**<font color='teal'> Currently, the missing values in the dataset are represented as zeros. Replace the zero values in the following columns ['Glucose','BloodPressure','SkinThickness','Insulin','BMI'] with nan .</font>**

```python
diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]  = diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0, np.nan)
```

**<font color='teal'> Plot histograms of each column. </font>**

```python
# Dictionary mapping column names to custom y-labels
y_labels = {
    'Pregnancies': 'Count of Individuals',
    'Glucose': 'Number of Measurements',
    'BloodPressure': 'Number of Readings',
    'SkinThickness': 'Number of Measurements',
    'Insulin': 'Number of Observations',
    'BMI': 'Number of Individuals',
    'DiabetesPedigreeFunction': 'Number of People',
    'Age': 'Count of People',
    'Outcome': 'Count of Cases'
}

# Dictionary mapping column names to x-label units
x_labels = {
    'Pregnancies': 'Number of Pregnancies',
    'Glucose': 'Glucose Level (mg/dL)',
    'BloodPressure': 'Blood Pressure (mmHg)',
    'SkinThickness': 'Skin Thickness (mm)',
    'Insulin': 'Insulin Level (U/mL)',
    'BMI': 'Body Mass Index (kg/m)',
    'DiabetesPedigreeFunction': 'Diabetes Pedigree Function (No Unit)',
    'Age': 'Age (years)',
    'Outcome': 'Diabetes Outcome (0 = No, 1 = Yes)'
}

# Loop through each column and plot histograms with customized labels
for col in diabetes_data.columns:
    plt.figure(figsize=(6, 4))  # Create a new figure for each histogram
    diabetes_data[col].hist(bins=15, edgecolor='black')  
    plt.title(f'Histogram of {col}')  # Title with column name
    plt.xlabel(x_labels.get(col, col))  # Get custom x-label with unit
    plt.ylabel(y_labels.get(col, 'Frequency'))  # Get custom y-label from dictionary
    plt.grid(True)  
    plt.show()  
```

#### Replace the zeros with mean and median values.

```python
diabetes_data.fillna({
    'Glucose': diabetes_data.Glucose.mean(),
    'BloodPressure': diabetes_data.BloodPressure.mean(),
    'SkinThickness': diabetes_data.SkinThickness.median(),
    'Insulin': diabetes_data.Insulin.median(),
    'BMI': diabetes_data.BMI.median()
}, inplace=True)
```

**<font color='teal'> Plot histograms of each column after replacing nan. </font>**

```python
# Dictionary mapping column names to custom y-labels
y_labels = {
    'Pregnancies': 'Count of Individuals',
    'Glucose': 'Number of Measurements',
    'BloodPressure': 'Number of Readings',
    'SkinThickness': 'Number of Measurements',
    'Insulin': 'Number of Observations',
    'BMI': 'Number of Individuals',
    'DiabetesPedigreeFunction': 'Number of People',
    'Age': 'Count of People',
    'Outcome': 'Count of Cases'
}

# Dictionary mapping column names to x-label units
x_labels = {
    'Pregnancies': 'Number of Pregnancies',
    'Glucose': 'Glucose Level (mg/dL)',
    'BloodPressure': 'Blood Pressure (mmHg)',
    'SkinThickness': 'Skin Thickness (mm)',
    'Insulin': 'Insulin Level (U/mL)',
    'BMI': 'Body Mass Index (kg/m)',
    'DiabetesPedigreeFunction': 'Diabetes Pedigree Function (No Unit)',
    'Age': 'Age (years)',
    'Outcome': 'Diabetes Outcome (0 = No, 1 = Yes)'
}

# Loop through each column and plot histograms with customized labels
for col in diabetes_data.columns:
    plt.figure(figsize=(6, 4))  # Create a new figure for each histogram
    diabetes_data[col].hist(bins=15, edgecolor='black')  
    plt.title(f'Histogram of {col}')  # Title with column name
    plt.xlabel(x_labels.get(col, col))  # Get custom x-label with unit
    plt.ylabel(y_labels.get(col, 'Frequency'))  # Get custom y-label from dictionary
    plt.grid(True)  
    plt.show()  
```

#### Plot the correlation matrix heatmap

```python
plt.figure(figsize=(12,10))
print('Correlation between various features')
p=sns.heatmap(diabetes_data.corr(), annot=True,cmap ='Blues')
```

**<font color='teal'> Define the `y` variable as the `Outcome` column.</font>**

```python
y = diabetes_data.Outcome
```

```python
X = diabetes_data.drop(columns=['Outcome'])
```

**<font color='teal'> Create a 70/30 train and test split. </font>**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

**<font color='teal'> Using Sklearn, standarize the magnitude of the features by scaling the values. </font>**

Note: Don't forget to fit() your scaler on X_train and then use that fitted scaler to transform() X_test. This is to avoid data leakage while you standardize your data.

```python
standardizer = StandardScaler()
```

```python
standardizer.fit(X_train)
standardizer.transform(X_test)
```

#### Using a range of neighbor values of 1-10, apply the KNearestNeighbor classifier to classify the the data.

```python
from sklearn.neighbors import KNeighborsClassifier


test_scores = []
train_scores = []

for i in range(1,10):

    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

**<font color='teal'> Print the train and test scores for each iteration.</font>**

```python
scores_df = pd.DataFrame({
    'n_neighbors': list(range(1, 10)),  # k values
    'Train Score': train_scores,
    'Test Score': test_scores
})
scores_df
```

**<font color='teal'> Identify the number of neighbors that resulted in the max score in the training dataset. </font>**

Number of neighbours = 1 resulted max score in the training dataset

**<font color='teal'> Identify the number of neighbors that resulted in the max score in the testing dataset. </font>**

Number of neighbours = 6 resulted max score in the testing dataset

Plot the train and test model performance by number of neighbors.

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(x=range(1,10), y=train_scores,marker='*', label='Train Score')
p = sns.lineplot(x=range(1,10), y=test_scores,marker='o', label='Test Score')
```

**<font color='teal'> Fit and score the best number of neighbors based on the plot. </font>**

```python
knn_best = KNeighborsClassifier(n_neighbors=6)

knn_best.fit(X_train, y_train)

train_score = knn_best.score(X_train, y_train)
test_score = knn_best.score(X_test, y_test)

print(f'Training Accuracy (k=6): {train_score:.4f}')
print(f'Test Accuracy (k=6): {test_score:.4f}')
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn_best.predict(X_test)
pl = confusion_matrix(y_test,y_pred)
```

**<font color='teal'> Plot the confusion matrix for the model fit above. </font>**

```python
plt.figure(figsize=(7, 5))
sns.heatmap(pl, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Diabetic', 'Diabetic'], yticklabels=['Not Diabetic', 'Diabetic'])
plt.title('Confusion Matrix for k-NN (k=6)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

**<font color='teal'> Print the classification report </font>**

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```

#### In the case of the K nearest neighbors algorithm, the K parameter is one of the most important parameters affecting the model performance.  The model performance isn't horrible, but what if we didn't consider a wide enough range of values in our neighbors for the KNN? An alternative to fitting a loop of models is to use a grid search to identify the proper number. It is common practice to use a grid search method for all adjustable parameters in any type of machine learning algorithm. First, you define the grid  aka the range of values  to test in the parameter being optimized, and then compare the model outcome performance based on the different values in the grid.

#### Run the code in the next cell to see how to implement the grid search method for identifying the best parameter value for the n_neighbors parameter. Notice the param_grid is the range value to test and we apply cross validation with five folds to score each possible value of n_neighbors.

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors':np.arange(1,50)}
knn = KNeighborsClassifier()
knn_cv= GridSearchCV(knn,param_grid, cv=5, n_jobs=-1)
knn_cv.fit(X_train,y_train)
```

#### Print the best score and best parameter for n_neighbors.

```python
print("Best Score:" + str(knn_cv.best_score_))
print("Best Parameters: " + str(knn_cv.best_params_))
```

```python
y_pred = knn_cv.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```

Here you can see that the ideal number of n_neighbors for this model is 14 based on the grid search performed. 

The ideal parameter has changed to 31.

**<font color='teal'> Now, following the KNN example, apply this grid search method to find the optimal number of estimators in a Randon Forest model.
</font>**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
param_grid = {'n_estimators': np.logspace(0, 3, 10, dtype=int)}
rf = RandomForestClassifier()
rf_cv = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1)
rf_cv.fit(X_train, y_train)
```

```python
print("Best Score:" + str(rf_cv.best_score_))
print("Best Parameters: " + str(rf_cv.best_params_))
```

```python
y_pred = rf_cv.predict(X_test)
report = classification_report(y_test, y_pred)
print(f"Best Parameters: {rf_cv.best_params_}\n")
print(f"Classification Report:\n{report}")
```

```python
param_grid = {'n_estimators': np.logspace(0, 4, 10, dtype=int)}
rf = RandomForestClassifier()
rf_cv = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1)
rf_cv.fit(X_train, y_train)
```

```python
print("Best Score:" + str(rf_cv.best_score_))
print("Best Parameters: " + str(rf_cv.best_params_))
```

```python
y_pred = rf_cv.predict(X_test)
report = classification_report(y_test, y_pred)
print(f"Best Parameters: {rf_cv.best_params_}\n")
print(f"Classification Report:\n{report}")
```

```python
param_grid = {'n_estimators': np.concatenate([np.arange(1, 101, 10), np.arange(100, 2001, 50)])}
# np.arange(1, 101, 10): Tests values from 1 to 100 with a step of 10 (coarse search).
# np.arange(100, 2001, 50): Tests values from 100 to 1000 with a step of 50 (finer search in a higher range).

rf = RandomForestClassifier()
rf_cv = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1)
rf_cv.fit(X_train, y_train)
```

```python
print("Best Score:" + str(rf_cv.best_score_))
print("Best Parameters: " + str(rf_cv.best_params_))
```

```python
y_pred = rf_cv.predict(X_test)
report = classification_report(y_test, y_pred)
print(f"Best Parameters: {rf_cv.best_params_}\n")
print(f"Classification Report:\n{report}")
```
